```python
import pandas as pd
file_path = "df_transposed_first_5000_income_statement99.csv"
df = pd.read_csv(file_path)
```

```python
df.head()
```

```python
df.count()
```

```python
# How many column have null value
df.isna().sum()
```

```python
# If droping all null value, it will give a dataframe with zero data since some of the column has zero data.
df.dropna()
```

```python
# If the column have more than 1461 null values, drop the column
# we do not want to drop the column of Total Revenue becuase we might be using this column in the future (1461 null values)
droped_column = []
for i in df.loc[:,]:
    if df[i].isnull().sum() > 1461:
        df.drop(i, axis=1, inplace=True)
        droped_column.append(i)
        
```

```python
df.isnull().sum()
```

```python
df.count()
```

```python
df.dropna(inplace=True)
```

```python
df.count()
```

```python
# Column that was dropped
droped_column
```
